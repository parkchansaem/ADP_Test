1-3 피쳐 엔지니어링을 통해 파생변수 1개를 생성하고 그 이유를 말하라

```python
# Age변수를 10대,20대,30대.. 등의 단위로 묶어 연령대간의 차이를 조금 더 포괄적으로 포함하는 파생변수를 생성
df['Age_new'] = df['Age'].apply(lambda x : x - x%10 )
df
```

2-1 앙상블을 제외한 분류 모델 3가지 구축 및 결과 비교 및 설명하라

```python
df.columns
```

```python
le_feature = ['Gender',
       'family_history_with_overweight', 'FAVC', 'FCVC',  'CAEC',
       'SMOKE', 'CH2O', 'SCC', 'FAF', 'TUE', 'CALC', 'MTRANS', 'NObeyesdad',
        ]
```

```python
df['family_history_with_overweight'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
df_n = df.copy()
le = LabelEncoder()
for feature in le_feature:
    df_n[feature] = le.fit_transform(df_n[feature])
x = df_n.drop(['id','NObeyesdad'],axis=1)
y = df_n['NObeyesdad']
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y)
```

```python
x_train.info()
```

```python
# 1. logisticRegression
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
train_score = LR.score(x_train,y_train)
test_score = LR.score(x_test,y_test)
print(f'LR score : {test_score}')
# 2. decisiontree
from sklearn.tree import DecisionTreeClassifier
Dtree = DecisionTreeClassifier()
Dtree.fit(x_train,y_train)
Dtree_train_score = Dtree.score(x_train,y_train)
Dtree_test_score = Dtree.score(x_test,y_test)
print(f'Tree score : {Dtree_test_score}')
# 3. SVC
from sklearn.svm import SVC
svc =SVC(probability=True)
svc.fit(x_train,y_train)
svc_train_score = svc.score(x_train,y_train)
svc_test_score = svc.score(x_test,y_test)
print(f'SVC score : {svc_test_score}')
```

```python
help(SVC)
```

-> 로지스틱회귀, 의사결정트리, 서포트벡터머신 등의 분류모델을 활용하여 비교하였다. 결과를 보았을 때 의사결정나무를 활용한 모델이 가장 높은 점수가 나왔다.
평가지표는 정확도를 사용하였고 다양한 평가지표를 함께 사용하여 모델을 비교하는 것도 방법이다.

2-2 2-1에서 사용한 모델 중 하나를 골라 그리드 서치를 통해서 파라미터 튜닝 및 분류 모델 성능 평가 (precision ,recall)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,precision_score,recall_score

# 그리드 서치에 사용할 파라미터 범위 설정
param_grid = {
    'C': [0.1, 1, 10],                # Regularization parameter
    'kernel': ['linear', 'rbf','poly'],      # Kernel 종류
    'gamma': ['scale', 'auto']        # 커널 함수에 대한 파라미터
}

grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(x_train,y_train)

y_pred = grid_search.predict(x_test)
# 4. Precision과 Recall 출력
print("Precision: ", precision_score(y_test, y_pred, average='weighted'))
print("Recall: ", recall_score(y_test, y_pred, average='weighted'))

# 5. Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

2-3 2-1의 3가지 모델을 soft voting을 이용하여 모델링 한 결과와 2-2과 비교하라

```python
from sklearn.ensemble import VotingClassifier
vo_clf = VotingClassifier(estimators=[('LR',LR),('Tree',Dtree),('SVC',svc)],voting='soft')
vo_clf.fit(x_train,y_train)

y_pred = vo_clf.predict(x_test)
#Precision과 Recall 출력
print("Precision: ", precision_score(y_test, y_pred, average='weighted'))
print("Recall: ", recall_score(y_test, y_pred, average='weighted'))

#Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
6-1 회귀 모델링 후 유의하지 않는변수 파악
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

lr = ols('Sales ~ TV + Radio + Newspaper',data = df)
result = lr.fit()
result.summary()
```

```python
6-2 변수 선택시 먼저 제거 될 변수 및 근거 제시
```

Newspaper의 p_value가 0.05를 넘으므로 통계적으로 유의하지 않은 변수라고 할 수 있다.

```python
6-3 VIF를 통한 다중공선성 진단
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
# 상수항 추가 (VIF 계산을 위한 전처리)
X_const = add_constant(df)  # 상수 항을 데이터에 추가

# 각 변수에 대한 VIF 계산
vif_data = pd.DataFrame()
vif_data['Variable'] = X_const.columns  # 변수 이름
vif_data['VIF'] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]  # VIF 계산

# 결과 출력
print(vif_data)
```
