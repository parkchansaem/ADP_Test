2. 차원축소
   2-1 차원축소 방법 2가지 이상을 비교하고 한 가지를 선택하시오
   요인분석: 요인분석은 수집된 많은 변수들이 있을 때, 유사한 항목들끼리 묶어서 처리하는 기법으로 모든 변수들 간의 관계를 분석함으로써 공통된 요인을 분석가가 묶어서 처리한다.
   주성분분석 : 서로 연관성이 높은 변수들의 선형결합으로 이루어진 주성분이라는 새로운 변수를 만들어 변수들을 요약하고 축소하는 기법.

   두가지 차원축소기법 중 요인분석은 분석가가 직접 공통 요인을 찾아야하므로 데이터에 대한 이해가 부족한 현재 상황에서는 주성분 분석을 선택하는 것이 올바르다.


```python
from sklearn.preprocessing import MinMaxScaler
df_scaled = df.copy()
scaler = MinMaxScaler()
sacled_amount = scaler.fit_transform(df[["Amount"]])
df_scaled["Amount"]=sacled_amount
```

```python
df_scaled_drop = df_scaled.drop(columns=["Time","Class"])
```

```python
df_scaled_drop
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=18)
```

```python
pca.fit(df_scaled_drop)
```

```python
print("PCA")
print("고유값 " , pca.singular_values_)
print("분산 설명력",pca.explained_variance_ratio_ )
```

```python
# Scree plot
plt.title("Scree Plot")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.plot(pca.explained_variance_ratio_,'o-')
```

```python
print("주성분 3: ", pca.explained_variance_ratio_[:3].sum()*100)
print("주성분 5: ", pca.explained_variance_ratio_[:5].sum()*100)
print("주성분 6: ", pca.explained_variance_ratio_[:6].sum()*100)
```

주성분 3개만하여도 80%를 차지하므로 3개로 진행 

3. 오버샘플링과 언더샘플링
   3-1. 오버 샘플링과 언더 샘플링 장단점 비교
   3-2. 분류분석 구현 및 위에서 선택한 샘플링 기법 중 2가지 이상의 알고리즘으로 모델을 비교하고 성능을 측정하시오
   3-3. 현재가지 전처리한 데이터를 통해 모델 수행후 결과를 분석

오버샘플링  
장점: 데이터의 손실없이 불균형 문제 해결  
단점: 데이터를 복제하며 특정 데이터에 대한 지나치게 학습하여 과적합의 위험이 있음  
언더 샘플링  
장점: 데이터가 줄어들어 복잡도가 낮아지고 학습속도가 빨라짐  
단점: 중요한 데이터가 삭제될수 있는 데이터 소실 문제 발생, 근복적으로 학습에 필요한 데이터가 부족하여 성능이 낮아질 수 있음  

```python
from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split
from sklearn.metrics import RocCurveDisplay
# svc_disp = RocCurveDisplay.from_estimator(svc, X_test, y_test)
df_pca=pca.transform(df_scaled_drop)
```

```python
pd.DataFrame(df_pca).iloc[:,:3]
```

```python
df['Class']
```

```python
# 원본
x_train,x_test,y_train,y_test = train_test_split(pd.DataFrame(df_pca).iloc[:,:3],df['Class'],test_size=0.3,random_state = 98)
clf = RandomForestClassifier(random_state=98)
clf.fit(x_train,y_train)
RocCurveDisplay.from_estimator(clf,x_test,y_test)
```

```python
# 오버샘플링
import imblearn. over_sampling
from imblearn.over_sampling import  RandomOverSampler

X_resampled, y_resampled = RandomOverSampler(random_state=0).fit_resample(x_train, y_train)

clf_re = RandomForestClassifier(random_state=98)
clf_re.fit(X_resampled,y_resampled)
RocCurveDisplay.from_estimator(clf_re,x_test,y_test)
```

```python
from imblearn.over_sampling import SMOTE
# 모델설정
sm = SMOTE( k_neighbors=5,random_state=2020)

X_resampled_over, y_resampled_over = sm.fit_resample(x_train,y_train)

clf_over = RandomForestClassifier(random_state=2020) 
clf_over.fit(X_resampled_over,y_resampled_over)

RocCurveDisplay.from_estimator(clf_over, x_test, y_test)
```

원본, 랜덤샘플링, smote샘플링 기법을 비교하였을 때 AUC=0.98로 유의미한 향상 효과를 보지 못하였다.
